# ***TCS*** v/s **INFOSYS**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
data = yf.download(['TCS.NS', 'INFY.NS'], start = '2022-01-01', end = '2024-01-01', auto_adjust=True, threads=False)['Close']
```

```python
data.head()
```

```python
returns = data.pct_change().dropna()
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(returns['TCS.NS'],
            returns['INFY.NS'],
            alpha=0.4,
            color='steelblue')
plt.xlabel('TCS Daily Return')
plt.ylabel('INFY Daily Return')
plt.title('Correlation: TCS vs INFY')

plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)

plt.show()
```

```python
#correlation:-
cor = returns['TCS.NS'].corr(returns['INFY.NS'])
print(f"Correlation between TCS and INFY: {cor:.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf

# Pair 1: Two IT stocks (expect high positive correlation)
data1 = yf.download(['TCS.NS', 'INFY.NS'],
                     start='2022-01-01',
                     end='2024-01-01')['Close']
returns1 = data1.pct_change().dropna()

# Pair 2: IT stock vs Gold (expect low/negative correlation)
data2 = yf.download(['TCS.NS', 'GOLDBEES.NS'],
                     start='2022-01-01',
                     end='2024-01-01')['Close']
returns2 = data2.pct_change().dropna()

# both scatter plots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.scatter(returns1['TCS.NS'], returns1['INFY.NS'],
            alpha=0.4, color='steelblue')
ax1.set_title(f"TCS vs INFY\nr = {returns1['TCS.NS'].corr(returns1['INFY.NS']):.2f}")
ax1.set_xlabel('TCS Return')
ax1.set_ylabel('INFY Return')
ax1.axhline(0, color='gray', linewidth=0.8)
ax1.axvline(0, color='gray', linewidth=0.8)

ax2.scatter(returns2['TCS.NS'], returns2['GOLDBEES.NS'],
            alpha=0.4, color='coral')
ax2.set_title(f"TCS vs Gold\nr = {returns2['TCS.NS'].corr(returns2['GOLDBEES.NS']):.2f}")
ax2.set_xlabel('TCS Return')
ax2.set_ylabel('Gold Return')
ax2.axhline(0, color='gray', linewidth=0.8)
ax2.axvline(0, color='gray', linewidth=0.8)

plt.tight_layout()
plt.show()
```
